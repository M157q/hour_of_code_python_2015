# Conda
Conda is a cross-platform package manger and enviroment manager.
Conda works on Windows, Mac, Linux.


##  Basci Usage


List conda environments

$ conda list -e

Create conda environment (**myenv** environment with python2.7 installed.)

$ conda create -n myenv python=2

Activate **myenv** environment

$ source activate myenv

$ conda install jupyter

[Get started with conda](http://pre.tir.tw/008/blog/output/conda-miniconda-python.html)

[Learn conda in 30 minutes](http://conda.pydata.org/docs/test-drive.html)

Remove conda environment
$ conda env remove -n myenv



# virtualenv, pyvenv

virtualenv is a tool to create isolated Python environments.

Install virtaualenv

$ pip install virtualenv

Usage

Create a new virtual environment **hello_env**

```
$ virtualenv-2.7 hello_env
New python executable in hello_env/bin/python2
Also creating executable in hello_env/bin/python
Installing setuptools, pip...done.
```

Activate this **hello_env** environment

```
$ source hello_env/bin/activate
```

In python3, we can use pyvenv instead of virtualenv.

The venv module is added after python3.3.

The venv module ( pyvenv) is used to create isolated environments.
Eash isolated environments can have its own independent set of installed Python packages.
